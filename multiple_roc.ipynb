```python
nbin=len(empf201903_ip2d_exB)
x_min=-20
x_max=50
xaxis=[x_min,x_max]
Δx=x_max-x_min
interval=Δx*(np.arange(0,nbin)/nbin)+x_min
print(len(interval)==nbin,nbin)

nbin_dl1=len(empf201903_dl1_exB)
xdl1_min=-8
xdl1_max=18
xdl1axis=[xdl1_min,xdl1_max]
Δx_dl1=xdl1_max-xdl1_min
interval_dl1=Δx_dl1*(np.arange(0,nbin_dl1)/nbin_dl1)+xdl1_min
print(len(interval_dl1)==nbin_dl1,nbin_dl1)
```

```python
ϵ=abs(x_max-x_min)/nbin
def integral(array,x):
    s=0
    i_min=int(round(nbin*(x[0]-x_min)/Δx))
    i_max=int(round(nbin*(x[1]-x_min)/Δx))
    for i in range(i_min,i_max):
        s=ϵ*array[i]+s
    return s

ϵ_dl1=abs(xdl1_max-xdl1_min)/nbin_dl1
def integral_dl1(array,x):
    s=0
    i_min=round(nbin_dl1*(x[0]-xdl1_min)/Δx_dl1)
    i_max=round(nbin_dl1*(x[1]-xdl1_min)/Δx_dl1)
    for i in range(i_min,i_max):
        s=ϵ_dl1*array[i]+s
    return s
```

```python
#### WORKS ONLY FOR MONOTONIC FUNCTIONS (INVERTIBLE FUNCTIONS) ####
#### FOR INCREASING F's, SET sgn=+1 ####
#### FOR DECREASING F's, SET sgn=-1 ####
def inverse(y,f,interval,sgn):
    if y<0.0 or y>1.0:
        return 'nan';
    ϵ=1e-7
    ϵ2=1e-7
    x_min=interval[0]
    x_max=interval[1]
    Δx=x_max-x_min
    q=1;n=0;
#    print(q, x_min, x_max)
    while abs(q)>ϵ and abs(Δx)>ϵ2:
#        x0=unif(x_min,x_max)    
        x0=0.5*(x_max+x_min)
            
        t=f(x0)
        q=t-y
        if sgn*q<=0:
            x_min=x0
        if sgn*q>0:
            x_max=x0
            
        Δx=x_max-x_min
        n=n+1
#        print(x0, t, sgn*q, [x_min, x_max])
    return x0,q,n
```

```python
cut_v=np.array([])
#cut_inverse=np.array([])

empf201903_rocip2d_rej=np.array([])
empf201903_rocip2d_rej_c=np.array([])

empf201903_rocip3d_rej=np.array([])
empf201903_rocip3d_rej_c=np.array([])

empf201903_rocrnnip_rej=np.array([])
empf201903_rocrnnip_rej_c=np.array([])

empf201903_rocsv1_rej=np.array([])
empf201903_rocsv1_rej_c=np.array([])

empf201903_rocjf_rej=np.array([])
empf201903_rocjf_rej_c=np.array([])

empf201903_rocdl1_rej=np.array([])
empf201903_rocdl1_rej_c=np.array([])

for x in x_ax:
    
    cut=inverse(x,lambda t: integral(empf201903_ip2d_exB/empf201903_N_exB,[t,x_max]),xaxis,-1)[0]    
#    cut_inverse=np.append(cut_inverse,integral(empf201903_ip2d_exB/empf201903_N_exB,[cut,x_max]))
    cut_v=np.append(cut_v,cut)
    y=integral(empf201903_ip2d_l/empf201903_N_l,[cut,x_max])
    y_c=integral(empf201903_ip2d_exC,[cut,x_max])/empf201903_N_exC
    empf201903_rocip2d_rej=np.append(empf201903_rocip2d_rej,y)
    empf201903_rocip2d_rej_c=np.append(empf201903_rocip2d_rej_c,y_c)
    
    cut=inverse(x,lambda t: integral(empf201903_ip3d_exB/empf201903_N_ip3dexB,[t,x_max]),xaxis,-1)[0]
    y=integral(empf201903_ip3d_l/empf201903_N_ip3dl,[cut,x_max])
    y_c=integral(empf201903_ip3d_exC,[cut,x_max])/empf201903_N_ip3dexC
    empf201903_rocip3d_rej=np.append(empf201903_rocip3d_rej,y)
    empf201903_rocip3d_rej_c=np.append(empf201903_rocip3d_rej_c,y_c)

    cut=inverse(x,lambda t: integral(empf201903_rnnip_exB/empf201903_N_rnnipexB,[t,x_max]),xaxis,-1)[0]
    y=integral(empf201903_rnnip_l,[cut,x_max])/empf201903_N_rnnipl
    y_c=integral(empf201903_rnnip_exC,[cut,x_max])/empf201903_N_rnnipexC
    empf201903_rocrnnip_rej=np.append(empf201903_rocrnnip_rej,y)
    empf201903_rocrnnip_rej_c=np.append(empf201903_rocrnnip_rej_c,y_c)  
    
    cut=inverse(x,lambda t: integral(empf201903_sv1_exB/empf201903_N_sv1exB,[t,x_max]),xaxis,-1)[0]
    y=integral(empf201903_sv1_l,[cut,x_max])/empf201903_N_sv1l
    y_c=integral(empf201903_sv1_exC,[cut,x_max])/empf201903_N_sv1exC
    empf201903_rocsv1_rej=np.append(empf201903_rocsv1_rej,y)
    empf201903_rocsv1_rej_c=np.append(empf201903_rocsv1_rej_c,y_c)     

    cut=inverse(x,lambda t: integral(empf201903_jf_exB/empf201903_N_jfexB,[t,x_max]),xaxis,-1)[0]
    y=integral(empf201903_jf_l,[cut,x_max])/empf201903_N_jfl
    y_c=integral(empf201903_jf_exC,[cut,x_max])/empf201903_N_jfexC
    empf201903_rocjf_rej=np.append(empf201903_rocjf_rej,y)
    empf201903_rocjf_rej_c=np.append(empf201903_rocjf_rej_c,y_c)     

    cut=inverse(x,lambda t: integral_dl1(empf201903_dl1_exB/empf201903_N_dl1exB,[t,xdl1_max]),xdl1axis,-1)[0]    
    y=integral_dl1(empf201903_dl1_l,[cut,xdl1_max])/empf201903_N_dl1l
    y_c=integral_dl1(empf201903_dl1_exC,[cut,xdl1_max])/empf201903_N_dl1exC   
    empf201903_rocdl1_rej=np.append(empf201903_rocdl1_rej,y)
    empf201903_rocdl1_rej_c=np.append(empf201903_rocdl1_rej_c,y_c)  
```

```python
rej_ip2d_leptonic_exact=1./empf201903_rocip2d_rej
rej_ip2d_c_leptonic_exact=1./empf201903_rocip2d_rej_c
rej_ip3d_leptonic_exact=1./empf201903_rocip3d_rej
rej_ip3d_c_leptonic_exact=1./empf201903_rocip3d_rej_c
rej_rnnip_leptonic_exact=1./empf201903_rocrnnip_rej
rej_rnnip_c_leptonic_exact=1./empf201903_rocrnnip_rej_c
rej_sv1_leptonic_exact=1./empf201903_rocsv1_rej
rej_sv1_c_leptonic_exact=1./empf201903_rocsv1_rej_c
rej_jf_leptonic_exact=1./empf201903_rocjf_rej
rej_jf_c_leptonic_exact=1./empf201903_rocjf_rej_c
rej_dl1_leptonic_exact=1./empf201903_rocdl1_rej
rej_dl1_c_leptonic_exact=1./empf201903_rocdl1_rej_c
```

```python
from scipy.signal import savgol_filter
pol_order=1
den=10
w_bin=int(len(rej_ip2d_leptonic_exact)/den)

if(int(w_bin%2)==0):
    q=+1
if(int(w_bin%2)!=0):
    q=-2
    
windows_size=w_bin+q
print(windows_size)
```

```python
rej_ip2d = savgol_filter(rej_ip2d_leptonic_exact, windows_size, pol_order)
rej_ip2d_c = savgol_filter(rej_ip2d_c_leptonic_exact, windows_size, pol_order)

rej_ip3d = savgol_filter(rej_ip3d_leptonic_exact, windows_size, pol_order)
rej_ip3d_c = savgol_filter(rej_ip3d_c_leptonic_exact, windows_size, pol_order)

rej_rnnip = savgol_filter(rej_rnnip_leptonic_exact, windows_size, pol_order)
rej_rnnip_c = savgol_filter(rej_rnnip_c_leptonic_exact, windows_size, pol_order)

rej_sv1 = savgol_filter(rej_sv1_leptonic_exact, windows_size, pol_order)
rej_sv1_c = savgol_filter(rej_sv1_c_leptonic_exact, windows_size, pol_order)

rej_jf = savgol_filter(rej_jf_leptonic_exact, windows_size, pol_order)
rej_jf_c = savgol_filter(rej_jf_c_leptonic_exact, windows_size, pol_order)

rej_dl1 = savgol_filter(rej_dl1_leptonic_exact, windows_size, pol_order)
rej_dl1_c = savgol_filter(rej_dl1_c_leptonic_exact, windows_size, pol_order)

mean=(rej_ip2d[-1]+rej_ip3d[-1]+rej_rnnip[-1]+rej_sv1[-1]+rej_dl1[-1])/5
sig=math.sqrt((rej_ip2d[-1]-mean)**2+(rej_ip3d[-1]-mean)**2+(rej_rnnip[-1]-mean)**2+(rej_sv1[-1]-mean)**2+(rej_dl1[-1]-mean)**2)/math.sqrt(5)
print(mean,sig)

mean_c=(rej_ip2d_c[-1]+rej_ip3d_c[-1]+rej_rnnip_c[-1]+rej_sv1_c[-1]+rej_dl1_c[-1])/5
sig_c=math.sqrt((rej_ip2d_c[-1]-mean_c)**2+(rej_ip3d_c[-1]-mean_c)**2+(rej_rnnip_c[-1]-mean_c)**2+(rej_sv1_c[-1]-mean_c)**2+(rej_dl1_c[-1]-mean_c)**2)/math.sqrt(5)
print(mean_c,sig_c)
```

```python
rej_ip2d_leptonic=np.zeros(len(rej_ip2d))
rej_ip2d_c_leptonic=np.zeros(len(rej_ip2d))
rej_ip3d_leptonic=np.zeros(len(rej_ip2d))
rej_ip3d_c_leptonic=np.zeros(len(rej_ip2d))
rej_rnnip_leptonic=np.zeros(len(rej_ip2d))
rej_rnnip_c_leptonic=np.zeros(len(rej_ip2d))
rej_sv1_leptonic=np.zeros(len(rej_ip2d))
rej_sv1_c_leptonic=np.zeros(len(rej_ip2d))
rej_jf_leptonic=np.zeros(len(rej_ip2d))
rej_jf_c_leptonic=np.zeros(len(rej_ip2d))
rej_dl1_leptonic=np.zeros(len(rej_ip2d))
rej_dl1_c_leptonic=np.zeros(len(rej_ip2d))


α=0.1
for i in range(0,len(rej_ip2d)):
    s=math.exp(α*(i-len(rej_ip2d)+1))
    rej_ip2d_leptonic[i]=rej_ip2d[i]-(rej_ip2d[-1]-1)*s
    rej_ip2d_c_leptonic[i]=rej_ip2d_c[i]-(rej_ip2d_c[-1]-1)*s
    rej_ip3d_leptonic[i]=rej_ip3d[i]-(rej_ip3d[-1]-1)*s
    rej_ip3d_c_leptonic[i]=rej_ip3d_c[i]-(rej_ip3d_c[-1]-1)*s
    rej_rnnip_leptonic[i]=rej_rnnip[i]-(rej_rnnip[-1]-1)*s
    rej_rnnip_c_leptonic[i]=rej_rnnip_c[i]-(rej_rnnip_c[-1]-1)*s
    rej_sv1_leptonic[i]=rej_sv1[i]-(rej_sv1[-1]-1)*s
    rej_sv1_c_leptonic[i]=rej_sv1_c[i]-(rej_sv1_c[-1]-1)*s    
    rej_jf_leptonic[i]=rej_jf[i]-(rej_jf[-1]-1)*s
    rej_jf_c_leptonic[i]=rej_jf_c[i]-(rej_jf_c[-1]-1)*s 
    rej_dl1_leptonic[i]=rej_dl1[i]-(rej_dl1[-1]-1)*s
    rej_dl1_c_leptonic[i]=rej_dl1_c[i]-(rej_dl1_c[-1]-1)*s
#print(rej_sv1_leptonic[990:],rej_sv1[990:])#,rej_ip2d[-1]-1)
```
